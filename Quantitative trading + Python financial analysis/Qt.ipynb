```python
import yfinance as yf
import pandas as pd

# List of stock symbols
stocks = ["AAPL", "MSFT", "INTC", "SPY", "^SPX","^VIX" ,"GDX", "GLD", "JPM","JNJ","KO","BA","PG", "GC=F"]

# Fetch data from 2010-01-01 to now and store in a DataFrame
dataframes = []
for stock in stocks:
    df = yf.download(stock, start="2010-01-01")[['Close']]
    df.rename(columns={'Close': stock}, inplace=True)
    dataframes.append(df)

# Merge all dataframes on the 'Date' column
merged_data = pd.concat(dataframes, axis=1)
```

```python
data =merged_data.dropna()
```

```python
windows = 10
```

```python
data['AAPL'].rolling(window=windows).min()
```

```python
data['min'] = data['AAPL'].rolling(window=windows).min()
data['max'] = data['AAPL'].rolling(window=windows).max()
data['mean'] = data['AAPL'].rolling(window=windows).mean()
data['std'] = data['AAPL'].rolling(window=windows).std()
```

```python
data.tail()
```

```python
data['a1']=data['AAPL'].rolling(window=30).mean()
```

```python
data['a2']=data['AAPL'].rolling(window=250).mean()
```

```python
data[['AAPL','a1','a2']].tail()
```

```python
data[['AAPL','a1','a2']].plot()
```

```python
data['positions']=np.where(data['a1']>data['a2'],1,-1)
```

```python
ax = rets.plot(kind='scatter', x='^SPX',y='^VIX')
ax.plot(rets['^SPX'], np.polyval(reg, rets['^SPX']), 'r')
```
